```python
np.where(marks>=50,"pass", "fail")
```

```python
np.where(marks>=50, marks, 0)
```

```python
data = {
    "city": ["chenna", "bangalore", "Mumbai", "Delhi", "hyderabad"],
    "rate": [5000, 5500, 30000, 6000, 5200]
}
```

```python
import pandas as pd
df = pd.DataFrame(data)
```

```python
df['clean_rate'] = np.where(df['rate']>10000, np.nan, df['rate'])
df
```
