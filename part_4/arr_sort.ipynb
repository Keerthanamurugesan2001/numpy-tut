```python
import numpy as np
series = np.array(['Dark', 'Wednesday', 'Money heist', 'Squid game', 'Stranger thing'])
rating = np.array([8.8, 8.1, 8.2, 8.0, 8.7])
```

```python
# ascending sort
```

```python
np.sort(rating)
```

```python
# descending sort
```

```python
np.sort(rating)[::-1]
```

```python
# argsort
```

```python
idx  =np.argsort(rating)[::-1]
```

```python
series[idx[0]]
```

```python
# partition
```

```python
np.partition(rating, -2)
```

```python
np.partition(rating, -2)[-2:]
```
